```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/NE_LINCOLN_NO_NA.csv")
data.columns
```

```python
# correlation matrix of gamma gross rate channels
corr = data.drop(columns = ['Unnamed: 0', "id", "location", "time", "doseEquivalent", "status"]).corr()
corr.style.background_gradient(cmap='coolwarm')
```

As seen from above correlation matrix, most of the gamma gross rate channels are correlated with each other with no two gamma rate gross channels having absolute correlation less than 0.5. The two channels that are the highest correlated are channel 3 and channel 5. The least correlated channels are gamma 8 and gamma 9. 

```python
# graphing gamma 3 vs gamma 5

plt.scatter(data['gamma3'], data['gamma5'], color = 'blue')
plt.xlabel("Gamma gross rate channel 3")
plt.ylabel("Gamma gross rate channel 5")
plt.show()
```

```python
# plotting least correlated channels (gamma8 vs gamma 9)
plt.close()
plt.scatter(data['gamma8'], data['gamma9'], color = 'green')
plt.xlabel("Gamma gross rate channel 8")
plt.ylabel("Gamma gross rate channel 9")
plt.show()
```
